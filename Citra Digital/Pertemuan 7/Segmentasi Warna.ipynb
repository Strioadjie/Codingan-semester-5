# Warna menggunakan k means clustering

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../photo/laptop adjie.jpg')

z = img.reshape((-1, 3))
z = np.float32(z)

k = 3
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
_, labels, centers = cv2.kmeans(z, k, 0, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

centers = np.uint8(centers)
segmented_img = centers[labels.flatten()]
segmented_img = segmented_img.reshape((img.shape))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Citra Asli")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title("segmentasi dengan k-means")
plt.imshow(cv2.cvtColor(segmented_img, cv2.COLOR_BGR2RGB))

plt.show()
```

# segmentasi warna dengan thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Baca citra
img = cv2.imread('../photo/laptop adjie.jpg') #Citra dalam format RGB

#1. Konversi ke ruang warna HSV
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

#2. Tentukan rentang warna untuk segmentasi
#Contoh: Segmentasi warna biru
lower_blue = np.array([100, 150, 50]) # Rentang bawah dalam HSV
upper_blue = np.array([130, 255, 255]) # Rentang atas dalam HSV

#3. Segmentasi dengan Thresholding
mask = cv2.inRange(hsv_img, lower_blue, upper_blue)

#4. Hasil segmentasi (aplikasikan mask ke citra asli)
segmented_img = cv2.bitwise_and(img, img, mask=mask)

#5. Tampilkan hasil
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Citra Asli")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title("Hasil Segmentasi Warna Biru")
plt.imshow(cv2.cvtColor(segmented_img, cv2.COLOR_BGR2RGB))

plt.show()
```

# sobel dan canny

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca citra
img = cv2.imread('../photo/laptop adjie.jpg', 0)

blurred_img = cv2.GaussianBlur(img, (5, 5), 0)

#2. Deteksi tepi dengan Sobel [cite: 47]
sobelx = cv2.Sobel(blurred_img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blurred_img, cv2.CV_64F, 0, 1, ksize=3)
sobel = np.hypot(sobelx, sobely)

canny_edges = cv2.Canny(blurred_img, 100, 200)

#4. Tampilkan hasil [cite: 61, 105]
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title("Citra Asli")
plt.imshow(img, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Deteksi Tepi Sobel")
plt.imshow(sobel, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Deteksi Tepi Canny")
plt.imshow(canny_edges, cmap='gray')

plt.show()
```
