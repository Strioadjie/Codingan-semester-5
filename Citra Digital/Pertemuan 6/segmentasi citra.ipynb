# active contour

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
from skimage.filters import gaussian
from skimage.segmentation import active_contour

image = data.astronaut()
image_gray = color.rgb2gray(image)
image_smooth = gaussian(image_gray, 3)
s = np.linspace(0, 2*np.pi, 400)
r = 0 + 100*np.sin(s)
c = 0 + 100*np.cos(s)
init = np.array([r, c]).T
snake = active_contour(image_smooth, init, alpha=0.015, beta=10, gamma=0.001)

fig, ax = plt.subplots(figsize=(7, 7))
ax.imshow(image_gray, cmap=plt.cm.gray)
ax.plot(init[:, 1], init[:, 0], '--r', lw=3, label='Inisialisasi kurva')
ax.plot(snake[:, 1], snake[:, 0], '-b', lw=3, label='active contour')
ax.legend()
plt.show()
```

# global thresholding

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('../photo/laptop adjie.jpg')
_, thresholded_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('gambar asli')

plt.subplot(1, 2, 2)
plt.imshow(thresholded_image, cmap='gray')
plt.title('hasil threshold global')
plt.show()
```

# adaptive thresholding

```python
import cv2

image = cv2.imread('../photo/laptop adjie.jpg', cv2.IMREAD_GRAYSCALE)

adaptive_thresh = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

plt.imshow(adaptive_thresh, cmap='gray')
plt.title('Adaptive Thresholding')
plt.show()
```

# otsu thresholding

```python
import cv2

image = cv2.imread('../photo/laptop adjie.jpg', cv2.IMREAD_GRAYSCALE)

_, otsu_thresh_image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(5, 5))
plt.imshow(otsu_thresh_image, cmap='gray')
plt.title('Otsu Thresholding')
plt.show()
```
