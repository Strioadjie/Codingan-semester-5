# implementasi morfologi pada citra biner

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar biner
img = cv2.imread('Image_1.jpg', cv2.IMREAD_COLOR_RGB)

# Membuat elemen struktural (misalnya, persegi 5x5)
kernel = np.ones((5,5), np.uint8)

# Melakukan berbagai operasi morfologi
dilation = cv2.dilate(img, kernel, iterations=1)
erosion = cv2.erode(img, kernel, iterations=1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

# Menampilkan hasil dari setiap operasi
titles = ['Original Image', 'Dilation', 'Erosion', 'Opening', 'Closing']
images = [img, dilation, erosion, opening, closing]

plt.figure(figsize=(10,5))
for i in range(5):
    plt.subplot(1,5,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```

# implementasi morfologi pada citra warna 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar warna
img = cv2.imread('Image_1.jpg')

# Memisahkan kanal warna R, G, B
b, g, r = cv2.split(img)

# Membuat elemen struktural (misalnya, persegi 5x5)
kernel = np.ones((5,5), np.uint8)

# Melakukan operasi morfologi pada setiap kanal warna
dilation_r = cv2.dilate(r, kernel, iterations=1)
dilation_g = cv2.dilate(g, kernel, iterations=1)
dilation_b = cv2.dilate(b, kernel, iterations=1)

# Menggabungkan kembali hasil operasi per kanal
dilated_img = cv2.merge((dilation_b, dilation_g, dilation_r))

# Menampilkan hasil
plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.subplot(1,2,2), plt.imshow(cv2.cvtColor(dilated_img, cv2.COLOR_BGR2RGB))
plt.title('Dilated Image')
plt.show()
```
