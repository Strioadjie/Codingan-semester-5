# ekstraksi ciri citra

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 

image = cv2.imread("Image_1.jpg", 0)
_, thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    area = cv2.contourArea(cnt)
    perimeter = cv2.arcLength(cnt, True)

    if perimeter > 0:
        circularity = 4* np.pi * (area / (perimeter ** 2))
        print(f"Area: {area}, Perimeter: {perimeter}, Circularity: {circularity}")
    else:
        print(f"Area: {area}, Perimeter: {perimeter}, Circularity: undefined (perimeter is zero)")

plt.imshow(thresh, cmap='gray')
plt.title("Threshold Image With Contours")
plt.show()
```
